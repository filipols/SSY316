```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
data = pd.read_csv('p4dataset2023.txt', sep=" ", header=None)

bases = data.iloc[:, 3:]
metadata = data.iloc[:, :3]
```

```python
mode_nucleobases = bases.mode().iloc[0] # computes the mode nucleobase for all columns
X_uncentered = (bases != mode_nucleobases).astype(int)
X = X_uncentered - X_uncentered.mean(axis=0)
```

```python
sklearn_pca = PCA()
X_PCA = sklearn_pca.fit_transform(X)
normalized_pca_components = sklearn_pca.components_
```

Q2

```python
num_classes = len(metadata.iloc[:, 2].unique())
classes = metadata.iloc[:, 2].unique()

size = 8

for i in range(0, X_PCA.shape[0]):
    population_class = metadata.iloc[i, 2]
    if population_class == classes[0]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='r',s=size)
    elif population_class == classes[1]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='b',s =size)
    elif population_class == classes[2]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='g',s =size)
    elif population_class == classes[3]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='y',s =size)
    elif population_class == classes[4]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='k',s =size)
    elif population_class == classes[5]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='m',s =size)
    elif population_class == classes[6]:
        plt.scatter(X_PCA[i, 0], X_PCA[i, 1], c='c',s =size)

plt.scatter([], [], c='r', label=classes[0])  # Empty plot for red class (class 0)
plt.scatter([], [], c='b', label=classes[1])  # Empty plot for blue class (class 1)
plt.scatter([], [], c='g', label=classes[2])  # Empty plot for green class (class 2)
plt.scatter([], [], c='y', label=classes[3])  # Empty plot for yellow class (class 3)
plt.scatter([], [], c='k', label=classes[4])  # Empty plot for black class (class 4)
plt.scatter([], [], c='m', label=classes[5])  # Empty plot for magenta class (class 5)
plt.scatter([], [], c='c', label=classes[6])  # Empty plot for cyan class (class 6)

plt.xlabel('v1')
plt.ylabel('v2')
plt.legend()
plt.show()
```
