```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
# Parameters
lambda_true = 2
number_of_samples = 2
alpha_init = 2
beta_init = 2

# Generate synthetic data from Poisson distribution
data = np.random.poisson(lambda_true, number_of_samples)

def calc_posterior_parameters(data, alpha_init, beta_init):
    alpha_post = alpha_init + np.sum(data)
    beta_post = beta_init + len(data)
    return alpha_post, beta_post

alpha_post, beta_post = calc_posterior_parameters(data, alpha_init, beta_init)

print('Posterior alpha:', alpha_post)
print('Posterior beta:', beta_post)


# Visualize the prior and posterior
x = np.linspace(0, 5, 1000)
prior = x**(alpha_init - 1) * np.exp(-beta_init * x)
posterior = x**(alpha_post - 1) * np.exp(-beta_post * x)
prior = prior / np.max(prior)
posterior = posterior / np.max(posterior)

plt.figure(figsize=(10, 5))
plt.plot(x, prior, label='Prior', color='blue')
plt.plot(x, posterior, label='Posterior', color='red')
plt.axvline(x=lambda_true, color = 'black', linestyle = '--', label='True lambda')
plt.xlabel('Lambda')
plt.ylabel('Probability density')
plt.title('Normalized prior and posterior distributions')
plt.legend()
```
