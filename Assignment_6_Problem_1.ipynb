```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
data = pd.read_csv('p4dataset2023.txt', sep=" ", header=None)

bases = data.iloc[:, 3:]
metadata = data.iloc[:, :3]
```

```python
mode_nucleobases = bases.mode().iloc[0] # computes the mode nucleobase for all columns
X_uncentered = (bases != mode_nucleobases).astype(int)
X = X_uncentered - X_uncentered.mean(axis=0)
```

```python
sklearn_pca = PCA()
X_PCA = sklearn_pca.fit_transform(X)
normalized_pca_components = sklearn_pca.components_
```
